# Multi-Provider Example: GyrB Target Evaluation

**Objective:** Demonstrate that the same code runs seamlessly across different LLM providers.

This notebook shows how the **Drug Discovery Pipeline** supports multiple LLM backends:
- Anthropic Direct API
- OpenAI (GPT-4 / GPT-3.5)
- Cohere
- Ollama (local)
- Together.ai
- AWS Bedrock

**Key point:** The code is identical; only the provider configuration changes.

## Step 1: Choose Your Provider

Run the cell below and select which LLM provider you want to use.

```python
# Uncomment ONE of the configurations below and run this cell

# ============================================================================
# OPTION 1: Anthropic Direct API (Recommended)
# ============================================================================
provider = "anthropic"
api_key = "sk-ant-YOUR_KEY_HERE"  # Or set ANTHROPIC_API_KEY env var
model = "claude-3-5-sonnet-20241022"
base_url = None

# ============================================================================
# OPTION 2: OpenAI (GPT-4 or GPT-3.5)
# ============================================================================
# provider = "openai"
# api_key = "sk-YOUR_KEY_HERE"  # Or set OPENAI_API_KEY env var
# model = "gpt-4-turbo"  # Or "gpt-4" or "gpt-3.5-turbo"
# base_url = None

# ============================================================================
# OPTION 3: Cohere
# ============================================================================
# provider = "cohere"
# api_key = "YOUR_COHERE_KEY"  # Or set COHERE_API_KEY env var
# model = "command-r-plus"
# base_url = None

# ============================================================================
# OPTION 4: Ollama (Local, Free)
# ============================================================================
# provider = "ollama"
# api_key = None  # Ollama doesn't require API key
# model = "llama2"  # Or "mistral", "neural-chat", etc.
# base_url = "http://localhost:11434/v1"  # Ensure Ollama is running: ollama serve

# ============================================================================
# OPTION 5: Together.ai
# ============================================================================
# provider = "together"
# api_key = "YOUR_TOGETHER_KEY"  # Or set TOGETHER_API_KEY env var
# model = "meta-llama/Llama-2-7b-chat-hf"
# base_url = None

# ============================================================================
# OPTION 6: AWS Bedrock
# ============================================================================
# provider = "bedrock"
# api_key = None  # Uses AWS CLI credentials
# model = "claude-3-5-sonnet-20241022"
# base_url = None

print(f"Provider: {provider}")
print(f"Model: {model}")
if base_url:
    print(f"Endpoint: {base_url}")
```

## Step 2: Initialize Client with Your Provider

```python
import sys
sys.path.insert(0, '..')

from src import DrugDiscoveryClient, APIConfig
from datetime import datetime
import json

# Create config with your chosen provider
config = APIConfig(
    provider=provider,
    api_key=api_key,
    model=model,
    base_url=base_url,
)

# Initialize client
try:
    client = DrugDiscoveryClient(config=config)
    print(f"✓ Client initialized successfully")
    print(f"  Provider: {client.config.provider}")
    print(f"  Model: {client.model_id}")
    if base_url:
        print(f"  Endpoint: {base_url}")
except Exception as e:
    print(f"✗ Failed to initialize client: {e}")
    print("\nTroubleshooting:")
    print(f"- Check your API key is correct for {provider}")
    print(f"- For Ollama, ensure 'ollama serve' is running")
    print(f"- Verify {model} is available for {provider}")
    raise
```

## Step 3: Define the Target (Same for All Providers)

```python
# Example: S. aureus GyrB (DNA Gyrase)
organism = "Staphylococcus aureus"
protein_name = "DNA gyrase subunit B"
protein_id = "GyrB"

print(f"Target: {organism} / {protein_name}")
print(f"ID: {protein_id}")
```

## Step 4: Run Evaluation (Same Code, Different Provider)

Notice: This is the **exact same code** that would work for any provider.

```python
print(f"\n[{datetime.now().strftime('%H:%M:%S')}] Running evaluation via {provider}...\n")

result = client.evaluate_target(
    organism=organism,
    protein_name=protein_name,
    protein_id=protein_id
)

print(f"[{datetime.now().strftime('%H:%M:%S')}] Evaluation complete.")
print("\n" + "="*70)
print(f"EVALUATION RESULT (via {provider.upper()})")
print("="*70 + "\n")
```

## Step 5: View Results

```python
# Print the full evaluation report
print(result["response"])
```

## Step 6: Save Results for Comparison

```python
# Save results for comparison across providers
output_file = f"gyrb_evaluation_{provider}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(output_file, "w") as f:
    json.dump(result, f, indent=2)

print(f"✓ Results saved to: {output_file}")
print(f"\nProvider: {provider}")
print(f"Model: {model}")
```

## Step 7: Compare Multiple Providers (Optional)

To compare results across providers:

1. Go back to **Step 1**
2. Uncomment a different provider option
3. Re-run **Step 2** (initialize client with new provider)
4. Re-run **Step 4** (evaluation)
5. Results will be saved with different filenames
6. Compare the outputs to see how different LLMs approach the same problem

This demonstrates the **provider-agnostic architecture** — your code doesn't change; only the LLM backend changes.
