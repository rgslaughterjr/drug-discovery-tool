# Workflow 2: Generate Validation Controls

**Objective:** Generate 10 known positive controls (binders) and 10 property-matched negative controls (decoys) to validate docking software.

**Inputs:**
- Organism
- Protein name
- PDB ID of the docking structure

**Outputs:**
- Table of 10 positive controls with IC50/Ki, SMILES, PubChem CIDs
- Table of 10 DUD-E-matched negative controls
- Literature references and validation notes

**Runtime:** ~90 seconds

## Step 1: Import and Configure

```python
import sys
sys.path.insert(0, '..')

from src import DrugDiscoveryClient
import json
from datetime import datetime

client = DrugDiscoveryClient()
print(f"✓ Client initialized ({client.config.provider})")
```

## Step 2: Define Target

```python
# === EDIT THESE ===

organism = "Staphylococcus aureus"
protein_name = "DNA gyrase subunit B"
pdb_id = "4P8O"  # From your target evaluation

print(f"Generating controls for: {organism} / {protein_name} ({pdb_id})")
```

## Step 3: Generate Controls

```python
print(f"\n[{datetime.now().strftime('%H:%M:%S')}] Generating controls...\n")

result = client.get_controls(
    organism=organism,
    protein_name=protein_name,
    pdb_id=pdb_id
)

print(f"[{datetime.now().strftime('%H:%M:%S')}] Generation complete.\n")
```

## Step 4: View Results

```python
print(result["response"])
```

## Step 5: Save and Export

```python
output_file = f"../examples/{protein_name}_controls_{pdb_id}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(output_file, "w") as f:
    json.dump(result, f, indent=2)

print(f"✓ Saved to: {output_file}")
print(f"\nNext step: Prepare screening campaign (Workflow 3)")
```
