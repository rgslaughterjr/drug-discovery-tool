# Workflow 3: Prepare Screening Campaign

**Objective:** Design a pharmacophore-based ChemBridge Diversity screening using ZINC20 database.

**Inputs:**
- Target organism, protein, PDB ID
- Binding mechanism (e.g., competitive NADPH inhibition)
- Docking software (optional)

**Outputs:**
- Ranked pharmacophore features
- Physicochemical filter cutoffs
- PAINS exclusion rules
- ZINC20 SMARTS query with expected library sizes
- Hit prioritization strategy

**Runtime:** ~120 seconds

## Step 1: Import and Configure

```python
import sys
sys.path.insert(0, '..')

from src import DrugDiscoveryClient
import json
from datetime import datetime

client = DrugDiscoveryClient()
print(f"✓ Client initialized ({client.config.provider})")
```

## Step 2: Define Screening Target

```python
# === EDIT THESE ===

organism = "Plasmodium falciparum"
protein_name = "Dihydrofolate reductase"
pdb_id = "1J3I"
mechanism = "Competitive NADPH inhibition"  # Describe the binding mechanism
docking_software = "Autodock Vina"  # Vina, DOCK6, Glide, rDock, or unspecified

print(f"Preparing screening campaign:")
print(f"  Target: {organism} / {protein_name}")
print(f"  PDB: {pdb_id}")
print(f"  Mechanism: {mechanism}")
print(f"  Software: {docking_software}")
```

## Step 3: Generate Screening Brief

```python
print(f"\n[{datetime.now().strftime('%H:%M:%S')}] Generating screening brief...\n")

result = client.prep_screening(
    organism=organism,
    protein_name=protein_name,
    pdb_id=pdb_id,
    mechanism=mechanism,
    docking_software=docking_software
)

print(f"[{datetime.now().strftime('%H:%M:%S')}] Complete.\n")
```

## Step 4: Review Results

```python
print(result["response"])
```

## Step 5: Save for Lab Use

```python
output_file = f"../examples/{protein_name}_screening_{pdb_id}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(output_file, "w") as f:
    json.dump(result, f, indent=2)

print(f"✓ Saved to: {output_file}")
print(f"\n→ Use the ZINC20 queries above to screen ChemBridge Diversity library")
print(f"→ After docking, proceed to Workflow 4: Analyze Hits")
```
