# Workflow 4: Analyze Virtual Screening Hits

**Objective:** Prioritize and rank compounds from virtual screening for wet-lab testing.

**Inputs:**
- Target protein name
- Number of compounds screened
- Docking score distribution (mean, SD, range)
- Positive control affinity data (optional)

**Outputs:**
- Score cutoff determination (anchored to positive controls)
- Murcko scaffold clustering
- PAINS re-filtering results
- Selectivity cross-check strategy
- Prioritized purchase list (top 10–20 compounds)
- Visual inspection checklist

**Runtime:** ~90 seconds

## Step 1: Import and Configure

```python
import sys
sys.path.insert(0, '..')

from src import DrugDiscoveryClient
import json
from datetime import datetime

client = DrugDiscoveryClient()
print(f"✓ Client initialized ({client.config.provider})")
```

## Step 2: Summarize Your Docking Results

```python
# === EDIT THESE WITH YOUR DOCKING RESULTS ===

protein_name = "S. aureus GyrB"
num_compounds = 180000  # Total compounds screened

# Docking score distribution (from your docking software output)
docking_scores_summary = """
Mean: -8.2 kcal/mol
SD: 1.1
Range: [-12.5, -4.3]
Median: -8.0
"""

# Positive control affinities (from your controls workflow)
positive_controls_affinity = """
SPR720 (known inhibitor): 0.063 µM (pIC50 7.2)
Clorobiocin (reference): 0.024 µM (pIC50 7.6)
Novobiocin (clinical): 0.15 µM (pIC50 6.8)
"""

print(f"Analyzing hits for: {protein_name}")
print(f"Compounds screened: {num_compounds:,}")
```

## Step 3: Analyze and Prioritize

```python
print(f"\n[{datetime.now().strftime('%H:%M:%S')}] Analyzing hits...\n")

result = client.analyze_hits(
    protein_name=protein_name,
    num_compounds=num_compounds,
    docking_scores_summary=docking_scores_summary,
    positive_controls_affinity=positive_controls_affinity
)

print(f"[{datetime.now().strftime('%H:%M:%S')}] Analysis complete.\n")
```

## Step 4: Review Prioritization

```python
print(result["response"])
```

## Step 5: Export for Purchase and Wet-Lab

```python
output_file = f"../examples/{protein_name}_hits_analysis_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(output_file, "w") as f:
    json.dump(result, f, indent=2)

print(f"✓ Saved to: {output_file}")
print(f"\n→ Use prioritized purchase list to order compounds")
print(f"→ Follow visual inspection checklist during manual review")
print(f"→ Proceed to biochemical assay validation")
```
